```python
def split_into_words(i):
  return (i.split(" "))
```

```python
# Preparing word count matrix format:

from sklearn.feature_extraction.text import CountVectorizer
bow=CountVectorizer(analyzer=split_into_words).fit(salary_train.Salary)
bow
```

```python
train_matrix=bow.transform(salary_train.Salary)
train_matrix.shape
```

```python
test_matrix=bow.transform(salary_test.Salary)
test_matrix.shape
```

```python
from sklearn.naive_bayes import MultinomialNB,GaussianNB, BernoulliNB
```

>> **1. Multinomial Naive Bayes**

```python
classifier_mb=MultinomialNB()
classifier_mb.fit(train_matrix.toarray(),salary_train.workclass)

train_pred_mb=classifier_mb.predict(train_matrix.toarray())
accuracy_train_mb=np.mean(train_pred_mb==salary_train.workclass)

test_pred_mb=classifier_mb.predict(test_matrix.toarray())
accuracy_test_mb=np.mean(test_pred_mb==salary_test.workclass)

print("Accuracy of trian data:",accuracy_train_mb)
print("Accuracy of test data: ",accuracy_test_mb)
```

>> **2. Multinomial Naive Bayes**

```python
classifier_g=GaussianNB()
classifier_g.fit(train_matrix.toarray(),salary_train.workclass)

train_pred_g=classifier_g.predict(train_matrix.toarray())
accuracy_train_g=np.mean(train_pred_g==salary_train.workclass)

test_pred_g=classifier_g.predict(test_matrix.toarray())
accuracy_test_g=np.mean(test_pred_g==salary_test.workclass)

print("Accuracy of trian data:",accuracy_train_g)
print("Accuracy of test data: ",accuracy_test_g)
```

>> **3. Bernoulli Naive Bayes**

```python
classifier_b=BernoulliNB()
classifier_b.fit(train_matrix.toarray(),salary_train.workclass)

train_pred_b=classifier_b.predict(train_matrix.toarray())
accuracy_train_b=np.mean(train_pred_b==salary_train.workclass)

test_pred_b=classifier_b.predict(test_matrix.toarray())
accuracy_test_b=np.mean(test_pred_b==salary_test.workclass)

print("Accuracy of trian data:",accuracy_train_b)
print("Accuracy of test data: ",accuracy_test_b)
```
